```python
import geopandas as gpd
import numpy as np
import pandas as pd
from bokeh.io import output_file, output_notebook, show
from bokeh.plotting import figure, ColumnDataSource
output_notebook()
```

```python
df = gpd.read_file('2018_Traffic_Flow_Counts/2018_Traffic_Flow_Counts.shp')
```

```python
df.head()
```

```python
# Extract street name, average weekday traffic, and lat/lon coordinate lists - from Kelsey

name_list = []
flow_list = []
lon_list = []
lat_list = []
for _, row in df.iterrows():
    name = row['STNAME_ORD']
    flow = row['AWDT']
    
    geo = row['geometry']
    try:
        lon_list.append([x for x,y in geo.coords])
        lat_list.append([y for x,y in geo.coords])
        name_list.append(name)
        flow_list.append(flow)
    except:
        for geom in geo.geoms:
            lon_list.append([x for x,y in geom.coords])
            lat_list.append([y for x,y in geom.coords])
            name_list.append(name)
            flow_list.append(flow)
            
bdf = pd.DataFrame({'name': name_list, 'flow': flow_list, 'lon': lon_list, 'lat': lat_list})
bdf.head()
```

```python
bdf = ColumnDataSource(bdf[["lat","lon"]])
```

```python
TOOLS = "pan,wheel_zoom,box_zoom,reset,hover,save" 
```

```python
Elevated = figure(title="Seattle", tools=TOOLS)   
Elevated.multi_line('lat', 'lon', source=bdf, color="black", line_width=2)
```

```python
show(Elevated)
```
